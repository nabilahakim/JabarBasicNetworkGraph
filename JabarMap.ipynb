## Menggunakan link map

```python
city = [
             'Bekasi',
             'Karawang',
             'Indramayu',
             'Cirebon',
             'Sumedang',
             'Bogor',
             'Cianjur',
             'Sukabumi',
             'Purwakarta',
             'Tasikmalaya',
             'Ciamis',
             'Banjar',
             'Bandung'
]

pos = [
            (2,4),
            (3,3.9),
            (9,3.8),
            (10, 3),
            (6,2.5),
            (1,3.1),
            (2.5,2),
            (1.5,1.5),
            (3.5,3.3),
            (8,1),
            (9,0.8),
            (10,0.5),
            (5,2)
]

jalan_utama_kota = [
                    ('Bekasi', 'Karawang'),
                    ('Bekasi', 'Bogor'),
                    ('Karawang', 'Purwakarta'),
                    ('Karawang', 'Indramayu'),
                    ('Bogor', 'Cianjur'),
                    ('Bogor', 'Sukabumi'),
                    ('Purwakarta', 'Indramayu'),
                    ('Purwakarta', 'Cianjur'),
                    ('Indramayu', 'Cirebon'),
                    ('Indramayu', 'Sumedang'),
                    ('Cianjur', 'Sukabumi'),
                    ('Cirebon', 'Sumedang'),
                    ('Tasikmalaya', 'Ciamis'),
                    ('Ciamis', 'Banjar'),
                    ('Bandung', 'Cianjur'),
                    ('Bandung', 'Purwakarta'),
                    ('Bandung', 'Sumedang'),
                    ('Bandung', 'Tasikmalaya')
]

display(len(city))
len(pos)
```

```python
G = nx.Graph()
for i in range(len(city)):
  G.add_node(city[i], pos=pos[i])

pos = nx.get_node_attributes(G, 'pos')
plt.figure(figsize=(10, 5))
nx.draw(G, pos, with_labels=True, node_color='yellow')
```

```python
G.add_edges_from(jalan_utama_kota)
plt.figure(figsize=(10, 5))
nx.draw(G, pos, with_labels=True, node_color='yellow', edge_color='red')
```

```python
nx.degree_centrality(G)
```

```python
print(f'Kota terpenting dari hasil degree: {", ".join([i for i in nx.degree_centrality(G) if nx.degree_centrality(G)[i] == max(nx.degree_centrality(G).values())])}')
```

```python
nx.betweenness_centrality(G)
```

```python
print(f'Kota terpenting dari hasil betweenness : {", ".join([i for i in nx.betweenness_centrality(G) if nx.betweenness_centrality(G)[i] == max(nx.betweenness_centrality(G).values())])}')
```
